```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('regular_season_totals_2010_2024.csv')
print(df.shape)
df.head()
```

```python
df.sort_values(by='GAME_DATE')
```

```python
df = df.sort_values(by='GAME_ID')
df.head(6)
```

```python
home_df = df[~df['MATCHUP'].str.contains('@')].copy()
away_df = df[df['MATCHUP'].str.contains('@')].copy()
```

```python
print(home_df.shape)
home_df.head(3)
```

```python
print(away_df.shape)
away_df.head(3)
```

```python
duplicates = ['SEASON_YEAR', 'GAME_DATE', 'MATCHUP', 'MIN', 'AVAILABLE_FLAG']
away_df = away_df.drop(columns=duplicates)
```

```python
duplicates.append('GAME_ID')
home_df = home_df.rename(columns=lambda col: f"H_{col}" if col not in duplicates else col)
away_df = away_df.rename(columns=lambda col: f"A_{col}" if col != 'GAME_ID' else col)
```

```python
merged = pd.merge(home_df, away_df, on='GAME_ID', how='inner')
merged.head(3)
```

```python
merged = merged.drop(columns=['A_WL'])

to_front = ['SEASON_YEAR',
            'GAME_ID', 
            'GAME_DATE', 
            'MIN', 
            'H_TEAM_ID', 
            'H_TEAM_ABBREVIATION', 
            'H_TEAM_NAME', 
            'A_TEAM_ID', 
            'A_TEAM_ABBREVIATION', 
            'A_TEAM_NAME']

stats = [col for col in merged if col not in to_front]

merged = merged[to_front + stats]
```

```python
merged.head(3)
```

```python
merged['SPREAD'] = merged.H_PTS - merged.A_PTS
sns.histplot(x='SPREAD', data=merged);
```

```python
merged['TOTAL'] = merged.H_PTS + merged.A_PTS
sns.histplot(x='TOTAL', data=merged);
```

```python
merged['OREB'] = merged.H_OREB + merged.A_OREB
sns.histplot(x='OREB', data=merged);
```

```python
basedf = merged.drop(columns=[col for col in merged.columns if "RANK" in col])
```

```python
basedf.info()
```

```python
advboxscores = []

for season in merged.SEASON_YEAR.unique():
    df = pd.read_csv(f"advboxscores/advboxscores{season}.csv")
    advboxscores.append(df)

advdf = pd.concat(advboxscores, ignore_index=True)
```

```python
advdf.rename(columns={"Match\xa0Up": "MATCHUP", "Game\xa0Date": "GAME_DATE"}, inplace=True)
```

```python
advdf['GAME_DATE'] = pd.to_datetime(advdf['GAME_DATE'])
basedf['GAME_DATE'] = pd.to_datetime(basedf['GAME_DATE'])
```

```python
advdf = advdf.drop(columns=['W/L', 'MIN'])
advdf.head()
```

```python
advhome = advdf[~advdf['MATCHUP'].str.contains('@')].copy()
advaway = advdf[advdf['MATCHUP'].str.contains('@')].copy()
```

```python
advhome.head(3)
```

```python
advaway.head(3)
```

```python
advaway['MATCHUP'] = advaway['MATCHUP'].apply(lambda x: f"{x.split(' @ ')[1]} vs. {x.split(' @ ')[0]}")
```

```python
merge_on = ["MATCHUP", "GAME_DATE"]

advhome = advhome.rename(columns=lambda col: f"H_{col}" if col not in merge_on else col)
advaway = advaway.rename(columns=lambda col: f"A_{col}" if col not in merge_on else col)

adv_merged = pd.merge(
    advhome,
    advaway,
    on=merge_on
)
```

```python
MASTERDF = pd.merge(
    basedf,
    adv_merged,
    on=merge_on,
    how="left"
)
```

```python
MASTERDF = MASTERDF.drop(columns=['H_Team', 'A_Team'])
print(MASTERDF.shape)
MASTERDF.head(3)
```

```python
MASTERDF.info()
```

```python
MASTERDF.to_csv("AdvancedGameStats.csv")
```
